```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import TensorDataset # 텐서데이터셋
from torch.utils.data import DataLoader #데이터 로더
```

텐서 데이터셋은 일반적으로 텐서를 입력으로 받는다. 텐서형태로 데이터를 정의한다.

```python
x_train  =  torch.FloatTensor([[73,  80,  75], 
                               [93,  88,  93], 
                               [89,  91,  90], 
                               [96,  98,  100],   
                               [73,  66,  70]])  
y_train  =  torch.FloatTensor([[152],  [185],  [180],  [196],  [142]])
```

이것을 TensorDataset의 입력으로 사용하고 dataset으로 저장한다.

```python
dataset = TensorDataset(x_train, y_train)
```

파이토치의 데이터셋을 만들엇다면 데이터 로더를 사용가능

데이터 로더는 기본적으로 다음 2개의 인자를 입력받는다. 하나는 데이터 셋, 그리고 미니배치의 크기이다.

shuffle은 Epoch마다 데이터셋을 섞어서 데이터가 학습되는 순서를 바꾼다.

```python
dataloader = DataLoader(dataset, batch_size = 2, shuffle = True)
```

```python
#모델과 옵티마이저 설계
model = nn.Linear(3, 1)
optimizer = torch.optim.SGD(model.parameters() , lr = 1e-5)
```

```python
epochs = 100
for epoch in range(epochs + 1):
  for batch_idx, samples in enumerate(dataloader):
    #print(batch_idx)
    #print(samples)
    x_train, y_train = samples

    # H(x) 계산
    pred = model(x_train)

    # cost
    cost = F.mse_loss(pred, y_train)

    # cost로 H(x) 계산
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print('Epoch {:4d}/{} Batch {}/{} Cost: {:.6f}'.format(epochs, epoch, batch_idx+1, len(dataloader), cost.item()))
```
